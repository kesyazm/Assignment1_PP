Membuat fungsi untuk melakukan preprocessing dan ekstraksi ciri TF-ID pada data teks tanpa menggunakan library.

```python
import math
```

Read Data (.txt)

```python
def read_txt_file(data):
    documents = []

    file = open(data, "r")

    for line in file:
        line = line.strip()
        if line != "":
            documents.append(line)

    file.close()

    return documents
```

Preprocessing

```python
def lowercase(text):
    return text.lower()
```

```python
def remove_punctuation(text):
    punctuation = ".,!?;:'\"()[]{}0123456789-"

    result = ""

    for char in text:
        if char not in punctuation:
            result += char

    return result
```

```python
def tokenize(text):
    words = []
    word = ""

    for char in text:
        if char != " ":
            word += char
        else:
            if word != "":
                words.append(word)
                word = ""

    if word != "":
        words.append(word)

    return words
```

```python
def remove_stopwords(tokens):

    stopwords = [
        "dan","yang","di","ke","dari","untuk","pada","adalah",
        "itu","ini","sebagai","dengan","atau","oleh","juga"
    ]

    filtered = []

    for word in tokens:
        if word not in stopwords:
            filtered.append(word)

    return filtered
```

```python
def preprocess(text):
    text = lowercase(text)
    text = remove_punctuation(text)
    tokens = tokenize(text)
    tokens = remove_stopwords(tokens)

    return tokens
```

Vocabulary

```python
def build_vocabulary(docs):
    vocab = []

    for doc in docs:
        for word in doc:
            if word not in vocab:
                vocab.append(word)

    return vocab
```

Term Frequency

```python
def term_frequency(doc, vocab):

    tf = []

    for term in vocab:

        count = 0

        for word in doc:
            if word == term:
                count += 1

        if len(doc) > 0:
            tf.append(count / len(doc))
        else:
            tf.append(0)

    return tf
```

Document Frequency

```python
def document_frequency(term, docs):
    count = 0

    for doc in docs:
        if term in doc:
            count += 1

    return count
```

IDF

```python
def inverse_document_frequency(vocab, docs):
    idf = []
    N = len(docs)

    for term in vocab:
        df = document_frequency(term, docs)
        value = math.log(N / (1 + df))
        idf.append(value)

    return idf
```

TF-IDF

```python
def compute_tfidf(documents):

    preprocessed_docs = []

    for text in documents:
        preprocessed_docs.append(preprocess(text))

    vocab = build_vocabulary(preprocessed_docs)

    idf = inverse_document_frequency(vocab, preprocessed_docs)

    tfidf_matrix = []

    for doc in preprocessed_docs:

        tf = term_frequency(doc, vocab)

        tfidf_vector = []

        for i in range(len(tf)):
            tfidf_vector.append(tf[i] * idf[i])

        tfidf_matrix.append(tfidf_vector)

    return vocab, tfidf_matrix, preprocessed_docs
```

Contoh Data

```python
documents = read_txt_file("data.txt")

vocab, tfidf_matrix, processed = compute_tfidf(documents)


print("=== PREPROCESSED DOCUMENTS ===")
for i,doc in enumerate(processed):
    print("Doc",i+1,":",doc)


print("\n=== VOCABULARY ===")
print(vocab)


print("\n=== TF-IDF MATRIX ===")

for i,row in enumerate(tfidf_matrix):

    print("\nDocument",i+1)

    for j,value in enumerate(row):

        if value > 0:
            print(vocab[j],":",round(value,5))
```
