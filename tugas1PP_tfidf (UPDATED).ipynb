# 2. Data Acquisition

Dataset yang digunakan dalam tugas ini berasal dari artikel Wikipedia mengenai Universitas Gadjah Mada yang dapat diakses melalui tautan berikut:

[Wikipedia: Universitas Gadjah Mada](https://id.wikipedia.org/wiki/Universitas_Gadjah_Mada)

Proses pengambilan data dilakukan secara manual dengan mengekstraksi beberapa kalimat informatif dari artikel tersebut. Kalimat yang diambil mencakup informasi mengenai sejarah, lokasi, struktur akademik, serta perkembangan Universitas Gadjah Mada.

Langkah-langkah pengumpulan dataset adalah sebagai berikut:
1.   Mengakses halaman Wikipedia Universitas Gadjah Mada.
2.   Memilih kalimat-kalimat informatif yang relevan dengan topik universitas.
3.   Membersihkan teks dari referensi, nomor kutipan, dan elemen yang tidak diperlukan.
4.   Menyimpan setiap kalimat sebagai satu dokumen dalam file teks (data.txt).

Dalam dataset ini, setiap baris pada file data.txt merepresentasikan satu dokumen teks. Total terdapat 20 dokumen yang digunakan sebagai corpus dalam proses analisis.

```python
with open("data.txt","r",encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip()]

corpus_raw = {}

for i,line in enumerate(lines):
    corpus_raw[f"dokumen_{i+1}"] = [line]

print("Jumlah dokumen:",len(corpus_raw))
```

# 3. Preprocessing

Tahap preprocessing bertujuan untuk membersihkan teks dan mempersiapkan data sebelum dilakukan ekstraksi fitur.

Tahapan preprocessing meliputi:
1. Parsing
2. Tokenization
3. Stopword Removal
4. Phrase Detection
5. Stemming
6. Lemmatization

##### **3.1. Parsing**

```python
def parsing(corpus_dict):

    doc_names=[]
    documents=[]
    sentences={}

    for name,kalimat_list in corpus_dict.items():
        doc_names.append(name)
        documents.append(" ".join(kalimat_list))
        sentences[name]=kalimat_list

    return doc_names,documents,sentences


doc_names,documents,sentences=parsing(corpus_raw)

print("TAHAP 1: PARSING\n")

for i,(name,doc) in enumerate(zip(doc_names,documents)):

    kalimat_list=sentences[name]

    print(f"Artikel {i+1}: {name}")
    print(f"Jumlah kalimat : {len(kalimat_list)}")
    print(f"Jumlah kata    : {len(doc.split())}\n")
```

##### **3.2. Tokenization/Lexical Analysis**

```python
def lexical_analysis(text):

    lowered=text.lower()

    cleaned=""

    for char in lowered:
        if 'a'<=char<='z':
            cleaned+=char
        else:
            cleaned+=' '

    tokens=cleaned.split()

    types=[]

    for w in tokens:
        if w not in types:
            types.append(w)

    return tokens,types


all_tokens_docs=[]

print("TAHAP 2: TOKENIZATION\n")

for i,name in enumerate(doc_names):

    tokens,types=lexical_analysis(documents[i])

    all_tokens_docs.append(tokens)

    print(f"Artikel {i+1}")
    print("Total Token:",len(tokens))
    print("Total Type :",len(types),"\n")
```

##### **3.3. Stop-word Removal**

```python
STOPWORDS={
"yang","dan","di","ke","dari","untuk","dengan","pada",
"adalah","ini","itu","atau","juga","dalam","oleh",
"sebagai","salah","satu","setelah","hingga"
}

def stopword_removal(tokens,stopwords):

    return [w for w in tokens if w not in stopwords]


all_filtered_docs=[]

print("TAHAP 3: STOPWORD REMOVAL\n")

for i,name in enumerate(doc_names):

    filtered=stopword_removal(all_tokens_docs[i],STOPWORDS)

    all_filtered_docs.append(filtered)

    print(f"Artikel {i+1}")
    print("Sebelum:",len(all_tokens_docs[i]))
    print("Sesudah:",len(filtered),"\n")
```

##### **3.4. Phrase Detection**

```python
import math

def compute_pmi(docs):

    unigram={}
    bigram={}

    total_uni=0
    total_bi=0

    for tokens in docs:

        for t in tokens:
            unigram[t]=unigram.get(t,0)+1
            total_uni+=1

        for i in range(len(tokens)-1):

            pair=(tokens[i],tokens[i+1])
            bigram[pair]=bigram.get(pair,0)+1
            total_bi+=1

    phrases=set()

    for (a,b),cnt in bigram.items():

        p_ab=cnt/total_bi
        p_a=unigram[a]/total_uni
        p_b=unigram[b]/total_uni

        pmi=math.log2(p_ab/(p_a*p_b))

        if pmi>=4 and cnt>=2:
            phrases.add((a,b))

    return phrases


phrases=compute_pmi(all_filtered_docs)

print("TAHAP 4: PHRASE DETECTION\n")
print("Jumlah bigram signifikan:",len(phrases))
```

##### **3.5. Stemming**

```python
PREFIXES=["di","ke","se","ber","ter","me","mem","men","meng","peng","pen","pem"]

def stem(word):

    for p in PREFIXES:
        if word.startswith(p) and len(word)>len(p)+2:
            return word[len(p):]

    return word


all_stemmed_docs=[]

print("TAHAP 5: STEMMING\n")

for tokens in all_filtered_docs:

    stemmed=[stem(w) for w in tokens]

    all_stemmed_docs.append(stemmed)
```

##### **3.6. Lemmatization**

```python
LEMMA={
"resm":"resmi",
"jad":"jadi"
}

all_lemmatized_docs=[]

print("TAHAP 6: LEMMATIZATION\n")

for tokens in all_stemmed_docs:

    lem=[LEMMA.get(w,w) for w in tokens]

    all_lemmatized_docs.append(lem)
```

# 4. Feature Extraction

##### **4.1. TF-IDF**

```python
def build_vocab(docs):

    vocab=[]

    for doc in docs:
        for w in doc:
            if w not in vocab:
                vocab.append(w)

    return vocab


def term_frequency(doc,vocab):

    tf=[]

    for t in vocab:
        tf.append(doc.count(t)/len(doc))

    return tf


def document_frequency(term,docs):

    return sum(1 for d in docs if term in d)


def compute_idf(vocab,docs):

    N=len(docs)

    idf=[]

    for t in vocab:
        df=document_frequency(t,docs)
        idf.append(math.log(N/(1+df)))

    return idf


vocab=build_vocab(all_lemmatized_docs)

idf=compute_idf(vocab,all_lemmatized_docs)

tfidf_matrix=[]

for doc in all_lemmatized_docs:

    tf=term_frequency(doc,vocab)

    vec=[tf[i]*idf[i] for i in range(len(vocab))]

    tfidf_matrix.append(vec)
```

# 5. Result

Berdasarkan hasil ekstraksi fitur menggunakan metode TF-IDF, setiap dokumen berhasil direpresentasikan dalam bentuk bobot numerik yang menunjukkan tingkat kepentingan suatu kata dalam dokumen tersebut. Nilai TF-IDF yang tinggi menunjukkan bahwa kata tersebut sering muncul pada dokumen tertentu tetapi jarang muncul pada dokumen lain, sehingga kata tersebut dianggap lebih representatif terhadap isi dokumen.

```python
print("=== TOP TF-IDF WORDS PER DOKUMEN ===\n")

for i,row in enumerate(tfidf_matrix):

    word_scores = []

    for j,value in enumerate(row):
        if value > 0:
            word_scores.append((vocab[j],value))

    word_scores.sort(key=lambda x:x[1],reverse=True)

    print(f"Dokumen {i+1}")

    for w,v in word_scores[:5]:
        print(f"{w:<15} {round(v,4)}")

    print()
```

# 6. Kesimpulan

Berdasarkan hasil ekstraksi fitur menggunakan metode TF-IDF (Term Frequency – Inverse Document Frequency), setiap dokumen berhasil direpresentasikan dalam bentuk bobot numerik yang menunjukkan tingkat kepentingan suatu kata dalam dokumen.

##### **Representasi Dokumen**

Metode TF-IDF mengubah teks yang semula berupa kalimat menjadi vektor numerik. Setiap dokumen direpresentasikan sebagai kumpulan kata dengan bobot tertentu yang mencerminkan tingkat kepentingannya dalam dokumen tersebut.

##### **Kata dengan Nilai TF-IDF Tinggi**

Hasil Top TF-IDF Words per Dokumen menunjukkan bahwa kata dengan nilai TF-IDF tertinggi biasanya berkaitan langsung dengan topik utama dokumen.

Beberapa contoh hasil yang diperoleh antara lain:
* Dokumen 1: kata negeri, kabupaten, dan sleman menunjukkan informasi mengenai lokasi Universitas Gadjah Mada.
* Dokumen 4: kata dunia dan peringkat menggambarkan informasi mengenai peringkat universitas di tingkat internasional.
* Dokumen 8: kata program, sarjana, magister, dan doktor berkaitan dengan jenjang pendidikan yang tersedia.
* Dokumen 16: kata sri, sultan, dan hamengkubuwono berkaitan dengan tokoh yang berperan dalam pendirian universitas.

##### **Fungsi TF-IDF dalam Analisis Teks**

Metode TF-IDF membantu mengidentifikasi kata-kata yang paling representatif dalam dokumen dengan cara:
* meningkatkan bobot kata yang sering muncul dalam dokumen tertentu, dan
* menurunkan bobot kata yang muncul di banyak dokumen.

Dengan demikian, kata yang bersifat umum memiliki bobot lebih kecil dibandingkan kata yang lebih spesifik terhadap dokumen.

##### **Kesimpulan Akhir**
Proses ekstraksi fitur menggunakan TF-IDF berhasil menghasilkan representasi numerik dari dokumen teks yang mampu menyoroti kata-kata paling penting dalam setiap dokumen. Representasi ini dapat digunakan dalam berbagai aplikasi text mining dan pattern recognition, seperti klasifikasi dokumen, clustering, dan analisis informasi berbasis teks.
